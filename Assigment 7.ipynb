```python
class Employee:
    def __init__(self, name, number, address, salary, post, year_of_joining, education, year_of_experience):
        self.name = name
        self.number = number
        self.address = address
        self.salary = salary
        self.post = post
        self.year_of_joining = year_of_joining
        self.education = education
        self.year_of_experience = year_of_experience

    def describe_employee(self):
        print(f"Name: {self.name}")
        print(f"Employee Number: {self.number}")
        print(f"Address: {self.address}")
        print(f"Salary: {self.salary}")
        print(f"Post: {self.post}")
        print(f"Year of Joining: {self.year_of_joining}")
        print(f"Education: {self.education}")
        print(f"Year of Experience: {self.year_of_experience}")

    def update_year_of_experience(self):
        current_year = 2024
        self.year_of_experience = current_year - self.year_of_joining

    def increase_salary(self, increment):
        self.salary += increment

# Creating an instance called employee101
employee101 = Employee("John Doe", 101, "123 Elm Street", 50000, "Engineer", 2018, "B.Tech", 6)

# Printing four attributes individually
print(employee101.name)
print(employee101.number)
print(employee101.address)
print(employee101.salary)

# Calling methods on employee101
employee101.describe_employee()
employee101.increase_salary(5000)
employee101.update_year_of_experience()

# Print updated details
print("\nUpdated Details:")
employee101.describe_employee()

# Creating 10 different instances and describing them
employees = [
    Employee("Alice Smith", 102, "456 Oak Avenue", 60000, "Manager", 2015, "MBA", 9),
    Employee("Bob Johnson", 103, "789 Pine Road", 55000, "Analyst", 2017, "M.Sc", 7),
    Employee("Charlie Brown", 104, "321 Maple Street", 52000, "Technician", 2019, "Diploma", 5),
    Employee("Diana Prince", 105, "654 Cedar Lane", 70000, "Director", 2012, "Ph.D.", 12),
    Employee("Eve Adams", 106, "987 Birch Boulevard", 45000, "Assistant", 2021, "B.A.", 3),
    Employee("Frank Castle", 107, "147 Spruce Street", 48000, "Coordinator", 2020, "B.Com", 4),
    Employee("Grace Lee", 108, "258 Willow Way", 53000, "Consultant", 2016, "MBA", 8),
    Employee("Hank Pym", 109, "369 Aspen Court", 62000, "Scientist", 2014, "M.Sc", 10),
    Employee("Ivy Green", 110, "741 Redwood Drive", 59000, "Advisor", 2013, "Ph.D.", 11),
    Employee("Jack Black", 111, "852 Dogwood Trail", 51000, "Developer", 2019, "B.Tech", 5)
]

# Describing each employee
for employee in employees:
    employee.describe_employee()
    print()  # Blank line between descriptions

# Performing some updates
employees[0].increase_salary(2000)
employees[1].update_year_of_experience()

# Print updated details for those employees
print("After Updates:")
employees[0].describe_employee()
employees[1].describe_employee()
```
